```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
```

```python
car_sales=pd.read_csv("car-sales.csv")
car_sales
```

```python
plt.bar(car_sales["Doors"],car_sales["Odometer (KM)"])
```

```python
sns.barplot(x=car_sales["Doors"],y=car_sales["Odometer (KM)"])
```

```python
sns.set(style="darkgrid");
sns.jointplot(x=car_sales["Doors"],y=car_sales["Odometer (KM)"]);
```

```python
heart_disease=pd.read_csv("heart-disease.csv")
heart_disease
```

```python
over_50=heart_disease[heart_disease["age"]>50]
heart=over_50.head(15)
heart
```

```python
sns.pairplot(heart_disease)
```

```python
#fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10));

#plot-1
sns.jointplot(x="age",y="chol",data=heart_disease,hue="target",palette="winter")
#plot-2
sns.jointplot(x="age",y="thalach",data=heart_disease,palette="spring",hue="target")
plt.show()
```

```python
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10));
#plot-1
sns.scatterplot(x="age",y="chol",data=heart_disease,ax=ax0,hue="target",palette="winter")
ax0.set(title="cholestrol and age")
ax0.axhline(heart_disease["chol"].mean(),linestyle="dotted")
ax0.axvline(heart_disease["age"].mean(),linestyle="dotted")



#plot-2
sns.scatterplot(x="age",y="thalach",data=heart_disease,ax=ax1,palette="spring",hue="target")
ax1.set(title="thalach and age")
fig.suptitle("HEART DISEASE DATA",fontsize=45)
ax1.axhline(heart_disease["thalach"].mean(),linestyle="dotted")
ax1.axvline(heart_disease["age"].mean(),linestyle="dotted")
plt.show()
```

```python
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(40,40));
#plot-1
sns.barplot(x="age",y="chol",data=over_50,ax=ax0,hue="target",palette="winter")
ax0.set(title="cholestrol and age")
ax0.axhline(heart_disease["chol"].mean(),linestyle="dotted")
ax0.axvline(heart_disease["age"].mean(),linestyle="dotted")
#ax0.set(xlim=[10,100])



#plot-2
sns.barplot(x="age",y="thalach",data=over_50,ax=ax1,palette="spring",hue="target")
ax1.set(title="thalach and age")
fig.suptitle("HEART DISEASE DATA",fontsize=45)
ax1.axhline(heart_disease["thalach"].mean(),linestyle="dotted")
ax1.axvline(heart_disease["age"].mean(),linestyle="dotted")
plt.show()
```
